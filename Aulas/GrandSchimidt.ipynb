# Algoritmo Grand Schimidt

```python
def proj(u,v):
    return (np.dot(v, u) / np.dot(u, u) * u)
```

```python
def ortonormalizacao(vetores):
    vetores_ort = []

    for v_k in vetores:
        soma = 0
        for u_i in vetores_ort:
            soma += proj(u_i, v_k)
        u_k = v_k - soma
        norm = np.linalg.norm(u_k)
        if np.abs(norm) < 1e-10:
            print("O conjunto Ã© LD")
            continue
        u_k = u_k / norm
        vetores_ort.append(u_k)
    return vetores_ort
```
