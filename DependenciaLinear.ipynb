# Dependencia e Independencia Linear


```python
v1_A = np.array([1, 0, -1])
v2_A = np.array([2, 2, 1])
v3_A = np.array([2, 6, 3])
v4_A = np.array([1, 5, 0])

conjunto_A = np.matrix([v1_A, v2_A, v3_A])
conjunto_B = np.matrix([v1_A, v2_A, v3_A, v4_A])

rank = np.linalg.matrix_rank(conjunto_A)
print(f"{conjunto_A.shape[0]} vetores, {rank} independentes. É L.I? {conjunto_A.shape[0] == rank}")
print(f"{conjunto_B.shape[0]} vetores, {rank} independentes. É L.I? {conjunto_B.shape[0] == rank}")
```
