# Descobrir Vetor a da Função Produto Interno

$ai = f(ei)$
<br>
$a = [f(e1), f(e2), ..., f(en)]$

```python
def x1_xn(vetor):
    return vetor[1] - vetor[-1]
```

```python
def vetor_a(f, n):
    a = np.array(n, dtype = "float")

    for i in range(n):
        e = np.zeros_like(a)
        e[i] = 1
        a[i] = f(e)
    
    return a
```

```python
vetor = np.array([1, 2, 4, 5, 6])
f_prod = lambda x, a: np.dot(x, vetor)
a = vetor_a(f_prod, 5)
```
